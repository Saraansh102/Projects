```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Israel-War.csv")
```

```python
df.head(20)
```

```python
df.info()
```

```python
df['date_of_event'] = pd.to_datetime(df['date_of_event'],format="mixed")
df['date_of_death'] = pd.to_datetime(df['date_of_death'],format="mixed")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df['age'].isnull().sum()
```

```python
#dropping the column notes
df.drop("notes" ,inplace=True ,axis=1)
```

```python
df.dropna(subset=["took_part_in_the_hostilities"],inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

```python
df["age"].mean().astype(int)
```

```python
df["age"].fillna(df["age"].mean().astype(int),inplace=True)
```

```python
df.head()
```

```python
df["gender"].mode()
```

```python
df["gender"].fillna(df["gender"].mode()[0],inplace=True)
```

```python
df["place_of_residence"].fillna("Unknown",inplace=True)
df["place_of_residence_district"].fillna("Unknown",inplace=True)
```

```python
df.isnull().sum()
```

```python
df["type_of_injury"].value_counts()
```

```python
df['type_of_injury'].fillna('Not Specified', inplace=True)
df['ammunition'].fillna('Not Specified', inplace=True)
```

```python
df.isnull().sum()
```

```python
x=df.shape
print(f" Dataset has {x[0]} rows and {x[1]} columns" )
```

```python
def showlabels(ax):
    for data in ax.containers:ax.bar_label(data)
```

## Data Analysis

```python
event=df.groupby(df["date_of_event"].dt.year).size()
print(event)
```

```python
plt.figure(figsize=(20,5))
line=plt.plot(event.index,event,marker="o",color='pink',mfc='red')
plt.title("Event Date")
plt.xlabel("Event")
plt.ylabel("Count")
plt.xticks(event.index)

plt.show
```

```python
plt.hist(df["age"], bins=30,edgecolor="Black")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid()
plt.show()
```

```python
count=df["citizenship"].value_counts()
print(count)
```

```python
plt.pie(count ,labels=count.index,autopct='%1.1f%%')
plt.title("Citizenship of People")
plt.show()
```

```python
event_loc=df["event_location"].value_counts().reset_index().head()
print(event_loc)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(event_loc["event_location"], event_loc["count"])
plt.xlabel("Event Location")
plt.ylabel("Count")
plt.title("Event Locations")
plt.xticks(rotation=45) 
plt.show()
```

```python
injuries=df["type_of_injury"].value_counts().reset_index()
print(injuries)
```

```python
plt.figure(figsize=(15,6))
plt.bar(injuries["type_of_injury"],injuries["count"])
plt.title("types of injuries")
plt.xlabel("injuries")
plt.ylabel("count")
plt.show()
```

```python
killed=df.groupby(["killed_by"])["gender"].value_counts().reset_index()
print(killed)
```

```python
plt.figure(figsize=(15,6))
plt.bar(killed["killed_by"] + ' - ' + killed["gender"],killed["count"])
plt.title("killed by")
plt.xlabel("killed by and gender")
plt.ylabel("count")
plt.legend()
plt.show()
```

```python
#another method
killed=df.groupby(["killed_by","gender"]).size().unstack()
print(killed)
```

```python
killed.plot(kind='bar', stacked=True, color=['skyblue', 'salmon'])
plt.xlabel("Killed_by")
plt.ylabel("Count")
plt.title("killed_by by Gender")
plt.xticks(rotation=0)
plt.legend(title="Gender")
plt.tight_layout()
plt.show()
```

```python
ammunation=df["ammunition"].value_counts().reset_index()
print(ammunation)
```

```python
plt.figure(figsize=(15,6))
plt.bar(ammunation["ammunition"],ammunation["count"])
plt.xticks(rotation=45)
plt.show()
```
